**Data Preprocessing:**

```python
# Preprocessing functions:

# for genres and keywords:
import ast
def extractGenres(allGenres):
  allGenres = ast.literal_eval(allGenres)   # converts string of lists into list
  result = []

  for genre in allGenres:
    result.append(genre['name'])

  return result

# for top 3 casts:
def getTopThreeCasts(allCasts):
  allCasts = ast.literal_eval(allCasts)
  result = []

  count = 0
  for cast in allCasts:
    if count == 3: break
    result.append(cast['name'])
    count += 1

  return result

# for Director from Crew:
def getDirector(allCrew):
  allCrew = ast.literal_eval(allCrew)
  result = []

  for crew in allCrew:
    if crew['job'].lower() == 'director':
      result.append(crew['name'])
      break

  return result
```

```python
movies['genres'] = movies['genres'].apply(extractGenres)
movies['keywords'] = movies['keywords'].apply(extractGenres)
movies['cast'] = movies['cast'].apply(getTopThreeCasts)
movies['crew'] = movies['crew'].apply(getDirector)
movies['description'] = movies['overview']
movies['overview'] = movies['overview'].apply(lambda x:x.split()) # converting overview into list so as to concatinate
```

```python
movies.head(3)
```

Removing spaces from tags

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
movies.head(3)
```

Making final Tag for word vectors:

```python
movies['tags'] =  movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] +  movies['crew']
```

**Final DataFrame:**

```python
newDf = movies[['id', 'title', 'tags', 'vote_average', 'description']]
newDf
```

```python
newDf.shape
```

```python
# convert tags list into string:
newDf.dtypes
newDf['tags'] = newDf['tags'].apply(lambda x: " ".join(x))
newDf['tags'] = newDf['tags'].apply(lambda x: x.lower())
newDf.head(2)
```

```python
newDf['tags'][0]
```

**Stemming:**

```python
# derives all the root words of a given text:
# stem('loving loved loves') = love ---> so as to avoid extra vectors
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    result = []
    for i in text.split():
        result.append(ps.stem(i))

    return " ".join(result)
```

```python
newDf['tags'] = newDf['tags'].apply(stem)
```

**Vectorization**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(newDf['tags']).toarray()  #converts scipy array to numpy array
vectors[0]
```

```python
cv.get_feature_names()  #all freq 5000 words
```

```python
newDf
```

**Cosine Similarity**

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)

similarity[0]   # tell similarity of this movie wrt other movies
```

**Recommender**

```python

# enumerate creates a indexing to each element present in similarity
def recommendMovies(movie):

  isPresent = movie in newDf['title'].values
  if isPresent == False:
    print("Sorry. This movie is not present in our database.")
    return

  id = newDf[newDf['title'] == movie].index[0]
  similarityScores = similarity[id]

  # map actual index and sort DESC on the basis of second element i.e. score
  result = sorted(list(enumerate(similarityScores)), reverse = True, key = lambda x: x[1])[1:6]

  for i in result:
    # i[0] is the id of present movie, get title from that
    print("Movie: ", newDf.iloc[i[0]].title)
    print("Overview: ", newDf.iloc[i[0]].description)
    print("Votes: ", newDf.iloc[i[0]].vote_average, end="\n\n")
```

```python
recommendMovies('Avatar')
```

```python
recommendMovies('Ba')
```
